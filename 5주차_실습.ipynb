```python
### 연도별 대출건수를 구한 후 대출건수를 기준으로 내림차순 정리하시오

df['발행년도'] = pd.to_numeric(df['발행년도'], errors='coerce')
loan_counts = df.groupby('발행년도')['대출건수'].sum().reset_index()
loan_counts = loan_counts.sort_values(by='대출건수', ascending=False).reset_index(drop=True)

print(loan_counts)
```

발행년도가 없는 출판사를 찾아, 출판사별 '발행년도 최빈값'으로 채워넣으시오
최종 df를 확인했을때 발행년도의 결측치가 0이여야함
단, 발행년도가 없을 시 랜덤년도 기입
단, 출판사가 없을 시 '출판사 없음' 입력

```python
no_year = df[pd.isnull(df['발행년도'])]
no_year
```

```python
#단, 출판사가 없을 시 '출판사 없음' 입력 후 df 확인

df['출판사'] = df['출판사'].fillna('출판사 없음')
df
```

```python
#발행년도가 없는 출판사를 찾아, 출판사별 '발행년도 최빈값'으로 채워넣으시오
#힌트! 발행년도가 없는 출판사의 리스트를 뽑아 for loop처리!
# 출판사가 없을 시 '출판사 없음' 입력
import random
import numpy as np

df['출판사'] = df['출판사'].fillna('출판사 없음')

# 각 출판사별 발행년도의 최빈값 계산
mode_year = df.groupby('출판사')['발행년도'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

# 발행년도가 없는 데이터의 인덱스 찾음
missing_year = df[df['발행년도'].isnull()].index

# 결측치 채움
for index in missing_year:
    publisher = df.loc[index, '출판사']
    mode_year = mode_year[publisher]
    
    # 최빈값이 없을 경우 랜덤한 년도로 채우기
    if pd.isna(mode_year):
        df.loc[index, '발행년도'] = random.randint(1900, 2023)  # 1900년에서 2023년 사이의 랜덤년도
    else:
        df.loc[index, '발행년도'] = mode_year

```
