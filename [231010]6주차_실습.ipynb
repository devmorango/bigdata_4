변동계수

```python
# 변동계수 : 데이터와의 상대적인 변동성을 나타내는 통계적 측정 지표
# 주로 표준편차와 평균사이의 관계를 사용해 계상함
# 주요 사용처: 다른 데이터 집합과의 비교, 서로 다름 데이터 집합의 상대적인 변동성을 비교할때 유용, 
# 변동계수가 작은 데이터 집합은 상대적으로 일정한 경향을 보이며, 변동계수가 큰 데이터 집합은 상대적으로 불안정하고 
# 큰 변동을 나타냄
# 단위가 다른 데이터 비교 : 서로 다른 측정 단위를 가지는 데이터를 비교할 때 사용,
# 변동계수는 상대적인 변동성을 단위에 독립적으로 나타내므로, 단위가 다른 데이터간에도 비교가 가능함
# 두개의 주식 포트폴리오가 있을때, 포트폴리오 A의 변동계수가 B의 변동계수보다 작다면, A는 비교적 안정적이고 덜 위험함
# 변동계수는 데이터의 상대적인 변동성을 빠르게 이해하고 비교하는데 도움을 주며 특히 금융, 경제, 공학 및 자연과학분야에 널리 사용됨
```

```python
##카이제곱 -->꽃받침 길이로 꽃을 분류하는데 쓸 수 있나?
##꽃받침길이간에 관계를 검증하는것, p-value <0.001(***): 유의미한 변수
###변수들을 활용해서 꽃을 분류한느 작업

from scipy.stats import chi2_contingency

# 'sepal length (cm)'을 quartile로 나눕니다.
iris_data['length_category'] = pd.qcut(iris_data['sepal length (cm)'], q=4)

# 교차 테이블 생성
cross_table = pd.crosstab(iris_data['class'], iris_data['length_category'])

# 카이제곱 검정
chi2, P, _, _ = chi2_contingency(cross_table)
print('chi2 결과:', chi2)
print('p-value:', P)
```
