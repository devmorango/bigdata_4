```python
##특정갑만 변경
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
#A컬럼의 모든값을 10으로 변경
df['A']=10
df
```

```python
##조건에 따라 값을 변경
##B 컬럼값이 5보다 큰 경우에, A를 20으로 변경
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df.loc[df['B']>5,'A']=20
df
```

```python
#인덱스 1의 행 A,B를 변경
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df.loc[1,['A','B']]=[50,60]
df
```

```python
#A 컬럼의 값이 50인 행의 B를 100으로 변경
df.loc[1,['A','B']]=[50,60]
```

```python
##apply를 활용한 값 업데이트
#'A'컬럼의 값에 5를 곱한 결과를 다시 C컬럼을 만들어서 저장

df['C']=df['A'].apply(lambda x:x*5)
df
```
